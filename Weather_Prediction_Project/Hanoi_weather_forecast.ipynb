```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
df_weather.info()
```

```python
labels = LabelEncoder()
df_weather['status_weather'] = labels.fit_transform(df_weather['status_weather'])
```

```python
df_weather['status_weather']
```

```python
features_cols = ['TS','T2M_MAX','T2M_MIN','RH2M','QV2M','PRECTOTCORR','WS10M','WS50M']
X = df_weather[features_cols]
Y = df_weather['status_weather']
```

```python
X_test = df_weather_test[features_cols]
```

```python
X
```

```python
data_tranformer = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
```

```python
preprocesor = ColumnTransformer(transformers=[
    ('num',data_tranformer,features_cols)   
])
```

```python
preprocesor.fit(X)
```

```python
X = preprocesor.transform(X) # X -> X_train, X_valid
```

```python
#using missing data in  X_test
X_test = preprocesor.transform(X_test)
```

```python
columns_name = ['TS','T2M_MAX','T2M_MIN','RH2M','QV2M','PRECTOTCORR','WS10M','WS50M']
X_test = pd.DataFrame(X_test,columns=columns_name)
X_test
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=2023)
```

```python
X_train.shape,X_val.shape,Y_train.shape,Y_val.shape,X_test.shape
```

## Using Native Bayes trainning model

```python
nb_model = GaussianNB()
nb_model.fit(X_train, Y_train)

y_pred = nb_model.predict(X_val)
```

```python
y_pred
```

```python
pd.DataFrame({'y': Y_val,'y_pred' : y_pred})
```

```python
acc0 = accuracy_score(Y_val, y_pred)
print(f"Accuracy score: {acc0}")
```

### Create visualization for all model with their Accuracy

```python
mylist=[]
mylist2=[]
mylist.append(acc0)
mylist2.append("Naive Bayes")
mylist.append(acc1)
mylist2.append("Logistic Regression")
mylist.append(acc2)
mylist2.append("SVM")
mylist.append(acc3)
mylist2.append("K_NN")
mylist.append(acc4)
mylist2.append("DTR")
mylist.append(acc5)
mylist2.append("RF")
mylist.append(acc6)
mylist2.append("XGBoost")
plt.rcParams['figure.figsize']=8,6
sns.set_style("darkgrid")
plt.figure(figsize=(22,8))
ax = sns.barplot(x=mylist2, y=mylist, palette = "mako", saturation =1.5)
plt.xlabel("Classification Models", fontsize = 20 )
plt.ylabel("Accuracy", fontsize = 20)
plt.title("Accuracy of different Classification Models", fontsize = 20)
plt.xticks(fontsize = 11, horizontalalignment = 'center', rotation = 8)
plt.yticks(fontsize = 13)
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.2%}', (x + width/2, y + height*1.02), ha='center', fontsize = 'x-large')
plt.show()
```
