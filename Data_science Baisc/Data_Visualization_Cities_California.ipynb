```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv("/content/drive/MyDrive/Python Tutorial/Dataset/california_cities.csv")
cities.head()

```

```python
cities["area_total_km2"].sort_values(ascending=False)
```

```python
# Extract latd(Vĩ độ) and longd(Kinh độ)

latd,longd = cities["latd"],cities["longd"]

latd[:10]
```

```python
longd[:10]
```

```python
plt.style.use("seaborn")
```

```python
#Plot using pylot API
populations , areas = cities["population_total"],cities["area_total_km2"]


plt.figure(figsize=(10,8))
plt.scatter(longd,latd, c=np.log10(populations), cmap="viridis",s=areas, linewidths=0,alpha=0.5);
plt.axis("equal")
plt.xlabel("longlatitude")
plt.ylabel("latitude")
plt.title("California USA")
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7)

area_range =[50,100,300,500,1000]

for area in area_range:
  plt.scatter([],[],s=area,c="k", label = str(area)+'km$^2$',alpha=0.4)

plt.legend(labelspacing=1.5,title="City area")
```
