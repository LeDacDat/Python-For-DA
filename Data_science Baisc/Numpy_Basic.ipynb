Dot product example


```python
#number of jars sold
np.random.seed(0)

sales_amounts = np.random.randint(20, size=(5,3))
sales_amounts
```

```python
#create weekly_sale dataframe
import pandas as pd

weekly_sales = pd.DataFrame(sales_amounts , index=["Mon","Tues","Wed","Thurs","Fri"],
                            columns=["Almon Butter", "Peanut Butter","Cashew Butter"])
weekly_sales
```

```python
#create  a price array
prices = np.array([10,8,12])
prices.shape
```

```python
butter_prices = pd.DataFrame(prices.reshape(1,3), index = ["Price"],
                             columns=["Almon Butter", "Peanut Butter","Cashew Butter"])
butter_prices
```

```python
weekly_sales.shape, butter_prices.shape
```

```python
total_price = weekly_sales.dot(butter_prices.T)
total_price
```

```python
weekly_sales["Total_Price"] = total_price
weekly_sales
```
