Pyplot API vs Object_Oriented(OO) API
* Pyplot API - quickly
* OO API -Advanced

**Pyplot API**

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x), color="blue",linestyle="dashed",label="Sin(x)");
plt.plot(x,np.cos(x), color="red", label="Cos(x)")
plt.title("Cos(X) and Sin(X)")
plt.axis("equal")


plt.xlabel("Biến X")
plt.ylabel("Sin(X)")
plt.legend()
plt.show()
```

**Object - Oriented API**

```python
#1 prepare data
x = np.array([1,2,3,4])
y = np.array([11,22,33,44])

#2 setup plot
fig , ax = plt.subplots(figsize=(5,5)) # figure size = width & height of the plot
ax.plot(x,y);
ax.set(title="A simple plot",
              xlabel="x-axis",
              ylabel="y-axis")
```

```python
#OO API
#preapare data
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = 1000*rng.randn(100)


fig,ax = plt.subplots()
img1 = ax.scatter(x,y,s=sizes,c=colors,cmap="viridis", alpha=0.5);
fig.colorbar(img1)
```

### Bar
* vertical chart
*horizontal


```python
#prepare data from a python dictionary

soft_drink_prices = {"Coca ": 12,
                     "Pessi ": 10,
                     "Spite ": 15}
fig,ax = plt.subplots()
ax.bar(soft_drink_prices.keys(),soft_drink_prices.values());
ax.set(title="Gia Do Uong Vinsmart",ylabel="Prices ($)")
```

```python
fig,ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()),list(soft_drink_prices.values()));
ax.set(title="Gia Do Uong Vinsmart",xlabel="Prices ($)",ylabel="Drinks")
```
